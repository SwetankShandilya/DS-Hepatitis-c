```python
pd.options.mode.chained_assignment = None

for feat in dfog.drop(['Unnamed: 0','Category','Sex'],axis=1):
    q3=dfog[feat].quantile(0.75)
    q1=dfog[feat].quantile(0.25)
    iqr=q3-q1
    up_lim=q3+(1.5*iqr)
    low_lim=q1-(1.5*iqr)
    dfog[feat]=np.where(
        dfog[feat]>up_lim,
        up_lim,
        np.where(
        dfog[feat]<low_lim,
        low_lim,
        dfog[feat]
    )
)
```

```python
cat_cols=dfog.select_dtypes('object').columns.tolist()
num_cols=dfog.select_dtypes(['int','float']).columns.tolist()
print(cat_cols)
print(num_cols)
```

```python
pd.options.mode.chained_assignment = None
# Scaling numeric columns using Mean  
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dfog[num_cols])
dfog[num_cols]=scaler.transform(dfog[num_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse=False)
sex_column_2d = dfog['Sex'].values.reshape(-1, 1)
ohe.fit(sex_column_2d)
sex_one_hot_encoded = ohe.transform(sex_column_2d)
df_sex_encoded = pd.DataFrame(sex_one_hot_encoded, columns=ohe.get_feature_names_out(['Sex']))
dfog = pd.concat([dfog, df_sex_encoded], axis=1)
```

```python
dfog=dfog.dropna()
dfog=dfog.drop('Sex',axis=1)
dfog
```

```python
dfog.isnull().sum()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
mod_par={
    
    'RFClas':{
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators':[5,10,20],
            'max_depth':[5,10,15]
        }
    },
    
    'LogReg':{
        'model': LogisticRegression(solver='liblinear'),
        'params':{
            'C':[1,5,10]
        }
    },
    
    'GBClass':{
        'model': GradientBoostingClassifier(),
        'params':{
        "n_estimators": [100, 200, 300],
        "learning_rate": [0.01, 0.1, 1],
        "max_depth": [3, 5, 10]
            
        }
    }
    
}
```

```python
train_y=dfog['Category']
train_x=dfog.drop('Category',axis=1)
```

```python
scores = []
for model_name, mp in mod_par.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=2, return_train_score=False)
    clf.fit(train_x, train_y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
pd.set_option('display.max_columns',3)
pd.set_option('display.max_colwidth', 100) 
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
